#### Ejercicio 1: ¿Cuántos títulos hay por cada tipo de contenido ('Movie' o 'TV Show')?

```python
#importaré primero la libreria de pandas y abriré el fichero de Netflix completo
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel("Df_Netflix_completo.xlsx",index_col=0)
df.head()
```

```python
#como nos pide cuantos títulos hay por cada tipo usaré la función count unida a un Groupby
pd_ejerc1 = pd.DataFrame(df.groupby(by="type")["title"].count())
print(f"en el tipo Movie hay {int(pd_ejerc1.loc["Movie","title"])} títulos")
print(f"en el tipo TV Show hay {int(pd_ejerc1.loc["TV Show","title"])} títulos")


```

#### Ejercicio 5: ¿Cuántos títulos se añadieron a Netflix cada año? Para resolver este ejercicio tendrás uqe convertir la columna `date_added` a formato fecha usando el método `pd.to_datetime`. [Aquí](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) tienes la documentación, leeala con detenimiento antes de usar el método. 

```python
#primero haré una nueva columna con el año de subida a Netflix en fución de la columna date_added
df["date_added"]= pd.to_datetime(df["date_added"])
df.dtypes
df["year_added"] =df["date_added"].dt.year
pd.DataFrame(df.groupby(by="year_added")["title"].count())

```

#### Ejercicio 10. ¿Cuál es el promedio de duración de las películas agrupadas por país?

```python
#primero convertiré la columna en númerica reemplazando los str "min" por nada.crearé una copia para no manchar el original solo con las películas
mascara10= df["type"]=="Movie"
df_pelis10 = df[mascara10].copy()
df_pelis10["duration"] = df_pelis10["duration"].str.replace(" min","").astype(float)
pd.DataFrame(df_pelis10.groupby(by="country")["duration"].mean().round(0))

```
